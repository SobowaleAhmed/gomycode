```python
# prompt: I want to create a simple embedding example using BERT

!pip install transformers sentence-transformers

from sentence_transformers import SentenceTransformer, util
import numpy as np

# Load the pre-trained Sentence Transformer model
model = SentenceTransformer('all-mpnet-base-v2')

# Example sentences
sentences = [
    "This is an example sentence.",
    "Each sentence is converted into an embedding.",
    "Sentences are passed as a list of string.",
    "The model creates an embedding for each sentence."
]

# Encode the sentences into embeddings
embeddings = model.encode(sentences)

# Calculate cosine similarity between the embeddings
cosine_scores = util.cos_sim(embeddings[0], embeddings[1:])
print("Cosine Similarities:", cosine_scores)

# Print the embeddings (optional)
# for i, embedding in enumerate(embeddings):
#   print(f"Sentence {i + 1}: {embedding}")


# Example of finding the most similar sentences
query_sentence = "This is a query sentence similar to the first one."
query_embedding = model.encode(query_sentence)
similarity_scores = util.cos_sim(query_embedding, embeddings)
most_similar_index = np.argmax(similarity_scores)
```

```python
print(f"Most similar sentence to '{query_sentence}' is: {sentences[most_similar_index]}")
```
