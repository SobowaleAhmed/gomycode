```python
def count_vowel(a):
  vowels = "aeiou"
  count = 0
  for i in a.lower():
    if i in vowels:
      count += 1
  print(count)



count_vowel("OLATUNBOSUN")
```

```python
def format_report(students):
    #Create a header for the report
    report = f"{'Name':<15}{'Age':<10}{'Score':<10}\n"
    report += "-" * 35 + "\n"
    # Add each student's information
    for student in students:
        name = student.get("name", "N/A")
        age = student.get("age", "N/A")
        score = student.get("score", "N/A")
        report += f"{name:<15}{age:<10}{score:<10}\n"
    return report


```

```python
# Example usage


data = [
    {"name": "Ahmed", "age": 26, "score": 85},
    {"name": "Pelumi", "age": 22, "score": 60},
    {"name": "David", "age": 19, "score": 78}
]

print(format_report(data))
```

```python
import math
def quad_equation(a:int, b:int, c:int):
  neg_b = -b
  square_b = b*b
  root_q = math.sqrt(square_b - (4*a*c))
  X1 = neg_b + root_q / 2
  X2 = neg_b - root_q / 2
  return f"x1 is {X1} and x2 is {X2}"
```
